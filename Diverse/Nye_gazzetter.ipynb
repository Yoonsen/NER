## Lag nye gazzetter

```python
import dhlab.nbtext as nb
```

Bygg URN-er med ikke-oversatt materiale - det skal utelukke spesielt sære navn

```python
urns = nb.book_corpus(limit=500000, trans='')
```

```python
len(urns)
```

```python
urns
```

Lag en liste med etternavn; det som står foran komma, og en liste av fornavn, første navn etter komma

```python
etternavn = list(map(lambda x: x.split(', ')[0], urns.author))
etternavn = list(set(etternavn))
```

```python
fornavn = list(map(lambda x: x.split(', ')[-1], urns.author))
fornavn = list(set(fornavn))
```

Ta bort tomme navn

```python
etternavn = [x for x in etternavn if x != '']
```

```python
etternavn
```

```python
[x.split()[0] for x in fornavn if x.split() != []]
```

Se om vi kan finne igjen navnene i listen:

```python
import re
```

```python
urns.author
```

```python
[x.group() for x in  (map(lambda x: re.match("{enavn}.*{fnavn}.*".format(enavn='Kagge', fnavn='Stein'), x), urns.author)) if x != None]
```

## Lagre etternavn og fornavn i mappe

```python
import json
```

```python
with open('NER_gazetter_norsk/etternavn.json', 'w', encoding="utf-8") as f:
    json.dump(etternavn, f)
```

```python
with open('NER_gazetter_norsk/fornavn.json', 'w', encoding="utf-8") as f:
    json.dump(fornavn, f)
```

```python
finn_navn = lambda fnavn, enavn, urns=urns: urns[urns.author.str.contains("{enavn}.*{fnavn}.*".format(enavn=enavn.capitalize(), fnavn=fnavn.capitalize()))]
```

```python
finn_navn('pet', 'zap')
```
