## Bygg et korpus

På vanlig måte, og inspiser.

```python
corpus = dh.Corpus(doctype="digibok", title = "reise i norge")
```

```python
corpus.corpus.head(5)
```

## Hant data for en tekst

Bruk URN og velg en modell fra spaCy. Modellene er satt opp som en integrert del av DH-lab. Nye modeller kan legges til, og vil synes gjennom `dh.Models()`.

```python
spacy = dh.Models()
```

```python
spacy.models
```

Anvend en av modellene på en URN. Det tar noen sekunder, men ikke håpløst lang tid for en enkelt tekst.

```python
geo = dh.GeoData(urn = corpus.corpus.urn[1], model= spacy.models[1])
```

Mulige stedsnavn ligger i attributtet `place_names`

```python
geo.place_names
```

Geolokasjonene beregnes ikke med en gang, men kan raskt finnes gjennom `add_geo_locations`, slik at dataene blir tilgjengelig for videre analyse gjennom attributtet `places`.

```python
geo.add_geo_locations()
```

```python
geo.places
```
